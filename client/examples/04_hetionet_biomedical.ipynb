# Hetionet Biomedical Knowledge Graph

Genes, diseases, compounds, symptoms, pathways, and more.

Requires `data/hetionet.json`.

```python
from graphbridge import GraphBridge
from graphbridge.datasets import load_hetionet
```

```python
gb = GraphBridge(frontend_port=5173)
gb.health()
```

```python
G = load_hetionet(max_genes=150, max_go_terms=200)
if G:
    print(f"Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")
```

```python
if G:
    from collections import Counter
    Counter(G.nodes[n]["label"] for n in G.nodes())
```

```python
if G:
    gb.load(G)
    gb.open()
```
