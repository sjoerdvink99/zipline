# Social Network

Synthetic social network with community structure (tech, sports, music).

```python
from graphbridge import GraphBridge
from graphbridge.datasets import make_social
```

```python
gb = GraphBridge(frontend_port=5173)
gb.health()
```

```python
G = make_social()
print(f"Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")
```

```python
from collections import Counter
Counter(G.nodes[n]["community"] for n in G.nodes())
```

```python
gb.load(G, node_label_attr="community")
gb.open()
```
